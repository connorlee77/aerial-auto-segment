```python
import os

import rasterio
from rasterio.plot import show

from utils.rasterio_utils import save_raster_preview_as_png, merge_rasters
```

```python
# Path to dirctory of tif tiles to merge. Directory structure should be:
# /.../{PLACE}/tiles/*.tif
DATA_PATH = '/data/microsoft_planetary_computer/dem'
SAVE_PREVIEW = True
```

```python
places = os.listdir(DATA_PATH)  
for place in places:
    merge_rasters(os.path.join(DATA_PATH, place, 'tiles'), os.path.join(DATA_PATH, place, 'mosaic'))
    if SAVE_PREVIEW:
        save_raster_preview_as_png(os.path.join(DATA_PATH, place, 'mosaic', 'mosaic.tiff'))
    if os.path.basename(DATA_PATH) == 'dem':
        merge_rasters(os.path.join(DATA_PATH, place, 'tiles_1m'), os.path.join(DATA_PATH, place, 'mosaic_1m'))
        if SAVE_PREVIEW:
            save_raster_preview_as_png(os.path.join(DATA_PATH, place, 'mosaic_1m', 'mosaic.tiff'))
```
